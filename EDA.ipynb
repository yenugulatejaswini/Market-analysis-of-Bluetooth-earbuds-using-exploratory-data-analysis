# Univariate Analysis

```python
df.columns
```

```python
def num_univariate(df, col, palette="viridis"):
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))

    # Histogram
    sns.histplot(
        df[col].dropna(),
        kde=True,
        ax=axes[0],
        color=sns.color_palette(palette)[2]
    )
    axes[0].set_title(f"{col} Distribution")

    # Boxplot
    sns.boxplot(
        x=df[col].dropna(),
        ax=axes[1],
        color=sns.color_palette(palette)[4]
    )
    axes[1].set_title(f"{col} Boxplot")

    plt.tight_layout()
    plt.show()

num_univariate(df, "Selling_Price", palette="magma")
num_univariate(df, "Rating", palette="Set2")
num_univariate(df, "battery_hours", palette="rocket")

```

# Categorical

```python
import seaborn as sns
import matplotlib.pyplot as plt

def cat_univariate(df, col, palette="Set2"):
    plt.figure(figsize=(8, 4))

    ax = sns.countplot(
        data=df,
        x=col,
        palette=palette,
        order=df[col].value_counts().index
    )

    total = len(df)
    for p in ax.patches:
        percent = f"{100 * p.get_height() / total:.1f}%"
        ax.annotate(
            percent,
            (p.get_x() + p.get_width() / 2., p.get_height()),
            ha="center",
            va="bottom"
        )

    plt.title(f"{col} Distribution")
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()

cat_cols = [
    'enc', 'anc', 'fast_charging', 'type_c', 'dual_pairing',
    'gaming_mode', 'low_latency', 'rgb', 'mic_count'
]

for col in cat_cols:
    cat_univariate(df, col, palette="Set2")

```

# Bivariate Analysis

# Numerical vs Numerical

```python
def num_vs_num_bivariate(df, x, y, palette="magma"):
    plt.figure(figsize=(7, 5))

    sns.regplot(
        data=df.dropna(subset=[x, y]),
        x=x,
        y=y,
        scatter_kws={
            "alpha": 0.6,
            "color": sns.color_palette(palette)[2]
        },
        line_kws={"color": "black"}
    )

    plt.title(f"{x} vs {y}")
    plt.xlabel(x)
    plt.ylabel(y)
    plt.tight_layout()
    plt.show()
num_vs_num_bivariate(df, "battery_hours", "Selling_Price")
num_vs_num_bivariate(df, "Rating_Count", "Rating", palette="viridis")
```

# categorical  vs Numerical

```python


def num_vs_cat_bivariate(df, num_col, cat_col, palette="Set3"):
    fig, axes = plt.subplots(1, 2, figsize=(14, 4))

    data = df.dropna(subset=[num_col, cat_col])

    # Boxplot
    sns.boxplot(
        data=data,
        x=cat_col,
        y=num_col,
        palette=palette,
        ax=axes[0]
    )
    axes[0].set_title(f"{num_col} by {cat_col}")
    axes[0].tick_params(axis="x", rotation=30)

    # Violin plot (with quartiles)
    sns.violinplot(
        data=data,
        x=cat_col,
        y=num_col,
        palette=palette,
        inner="quartile",
        ax=axes[1]
    )
    axes[1].set_title(f"{num_col} Distribution by {cat_col}")
    axes[1].tick_params(axis="x", rotation=30)

    plt.tight_layout()
    plt.show()
num_vs_cat_bivariate(df, "Selling_Price", "gaming_mode")
num_vs_cat_bivariate(df, "Rating", "enc")
num_vs_cat_bivariate(df, "battery_hours", "fast_charging")

```

# Categorical vs Categorical

```python
def cat_vs_cat_bivariate(df, col1, col2, palette="tab10"):
    plt.figure(figsize=(8, 5))

    data = df.dropna(subset=[col1, col2])

    ax = sns.countplot(
        data=data,
        x=col1,
        hue=col2,
        palette=palette
    )

    plt.title(f"{col1} vs {col2}")
    plt.xticks(rotation=30)
    plt.legend(title=col2)
    plt.tight_layout()
    plt.show()
cat_vs_cat_bivariate(df, "gaming_mode", "low_latency")
cat_vs_cat_bivariate(df, "enc", "anc")
cat_vs_cat_bivariate(df, "fast_charging", "type_c")

```
