# Data cleaning (missing values)

```python
cols=['Rating', 'Rating_Count', 'Selling_Price',  'battery_hours']
for i in cols:
    print(f"{i},{df[i].skew()}")
```

```python
cols = ["Rating", "Rating_Count", "Selling_Price", "battery_hours"]

for col in cols:
    df[col].fillna(df[col].median(), inplace=True)
df["brand"].fillna("Unknown",inplace=True)
```
