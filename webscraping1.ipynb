```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import csv

options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled")

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

base_url = "https://www.flipkart.com/search?q=boat%20earbuds&page="
driver.get(base_url + "1")
time.sleep(5)

# Close login popup
try:
    driver.find_element(By.XPATH, "//button[contains(text(),'âœ•')]").click()
except:
    pass

products = []
MAX_PRODUCTS = 1000
page = 1

while len(products) < MAX_PRODUCTS:
    print(f"Scraping page {page}...")
    time.sleep(4)

    cards = driver.find_elements(By.CSS_SELECTOR, "a.plpbg")

    for card in cards:
        try:
            name = card.find_element(By.CSS_SELECTOR, "a[title]").text
        except:
            name = "N/A"

        try:
            color = card.find_element(By.CSS_SELECTOR, "div.U_GKrr").text
        except:
            color = "N/A"

        try:
            rating = card.find_element(By.CSS_SELECTOR, "div.MkI5F6").text
        except:
            rating = "N/A"

        try:
            rating_count = card.find_element(By.CSS_SELECTOR, "span.PvNbMB").text
        except:
            rating_count = "N/A"

        try:
            price = card.find_element(By.CSS_SELECTOR, "div.H3PBw").text
        except:
            price = "N/A"

        try:
            original_price = card.find_element(By.CSS_SELECTOR, "div.kRYcDn").text
        except:
            original_price = "N/A"

        try:
            discount = card.find_element(By.CSS_SELECTOR, "div.HQe8jr").text
        except:
            discount = "N/A"

        try:
            url = card.get_attribute("href")
        except:
            url = "N/A"

        products.append({
            "name": name,
            "color": color,
            "rating": rating,
            "rating_count": rating_count,
            "price": price,
            "original_price": original_price,
            "discount": discount,
            "url": url
        })

        if len(products) >= MAX_PRODUCTS:
            break

    page += 1
    driver.get(base_url + str(page))

driver.quit()

print(f"Total scraped: {len(products)}")

# Save CSV
with open("boat_earbuds_1000_flipkart.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(
        f,
        fieldnames=[
            "name", "color", "rating", "rating_count",
            "price", "original_price", "discount", "url"
        ]
    )
    writer.writeheader()
    writer.writerows(products)
```
